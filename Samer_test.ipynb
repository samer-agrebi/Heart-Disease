```python
data=pd.get_dummies(data,drop_first=True)
data.head()
```

#fitting into x and y

```python
x=data.drop("target",axis=1)
y=data["target"]
```

#training and testing the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1234,test_size=0.3,stratify=y)
```

#classification model

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=1234)
rfc.fit(x_train,y_train)
```

#predicting the data

```python
y_predict=rfc.predict(x_test)
y_predict
```

#calculating r-square

```python
score=rfc.score(x_test,y_test)
score
```

#drawing confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
cm
```

#drawing accuracy_score


```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_predict)
```

#drawing classification_report


```python
classification_report(y_test,y_predict)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split

```

```python
DT = DecisionTreeClassifier()
DT.fit(x_train,y_train)
predict_test = DT.predict(x_test)
print(DT.fit(x_train,y_train).tree_.node_count)
```

```python
print("Binary Output DecisionTreeClassifier",metrics.accuracy_score(predict_test,y_test),'\n')
print(metrics.classification_report(y_test, predict_test))
```
